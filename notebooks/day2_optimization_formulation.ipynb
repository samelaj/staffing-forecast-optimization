Hourly Optimization for a single day
- We will optimize bike avialiability across 24 hours for single day

Decision Variable: $x_h$ = number of bikes allocated at hour h where h = 1,2,...,24

Known Quantities:<br>
                $d_h$ = average demand at hour h <br>
                $C$ = total bike capacity

Optimization Problem:
<br>
                    $max_{x_h}$ $\sum_{h=1}^{24} min(x_h,d_h)$  <br> subject to: <br>$\sum_{h=1}^{24} x_h \le C$ <br> $x_h \ge 0$

Converting to solvable problem
$y_h$ = fulfilled demand at hour h

$max \sum_h y_h$
<br>
subject to:
<br>
$y_h \le x_h $
<br>
$y_h \le d_h $
<br>
$\sum_h x_h \le C $
<br>
$x_h, y_h \ge 0 $
