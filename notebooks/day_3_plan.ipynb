How will LP become more realistic
- Current LP just maximizes demand $\sum_{h} y_h$ therefore no penalty for when bikes are allocated
Error in current LP
- The intial LP solution concentrated bike allocation in late hours de=ue to the absence of temporal structure
Updated LP (Adding new variable and constraint)<br>
h = 0,1,2,...,23 (hours) <br>
$x_h$ (number of bikes allocated at hour h) <br>
$y_h$ (number of fulfilled rentals at hour h) <br>
$d_h$ (demand at hour h) <br>
$C$ (total bike capacity) <br>
$R$ (rental duration in hours) <br>

Objective: $max \sum_{h=0}^{23}y_h$ <br>
Subject to <br>
$y_h \le x_h$ <br>
$y_h \le d_h$ <br>
$\sum_{k=0}^{R-1} x_{h-k} \le C$ <br>
$x_h \ge 0, y_h \ge 0$ <br>
