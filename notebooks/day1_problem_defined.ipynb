What does demand mean?
 - Demand is the amount of registered users using the bike rental service during a given period

What is the time scale?
- Hourly Demand Modeling to capture intra-day usage patterns
- Daily aggregation to study longer-term trends and seasonality

What are constraints?
- Bike supply constraint: total rentals cannot exceed the number of available bikes
- Weather constraint: demand is influenced by a weather severity index (weathersit, scored 1-4)
- Calender effects: De

What are we optimizing
- The total number of registered users
